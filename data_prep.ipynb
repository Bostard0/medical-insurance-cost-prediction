```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
insurance_data = pd.read_csv("insurance.csv")
```

```python
insurance_data.info()
```

```python
insurance_data.describe()
```

```python
insurance_data['age'].hist(bins = 20)
plt.xlabel('age')
plt.ylabel('Frequency')
plt.title('Histogram of age')
plt.show()
```

```python
sex_values = insurance_data['sex'].value_counts()
sex_values.plot(kind='bar')
plt.xlabel('sex')
plt.ylabel('count')
plt.title('Distribution of sex')
plt.show()
```

```python
sex_values
```

```python
insurance_data['bmi'].hist(bins = 50)
plt.xlabel('bmi')
plt.ylabel('Frequency')
plt.title('Histogram of bmi (ideal values are between 18.5 and 24.9')
plt.show()
```

```python
children_values = insurance_data['children'].value_counts()
children_values.plot(kind='bar')
plt.xlabel('children')
plt.ylabel('count')
plt.title('Distribution of children')
plt.show()
```

```python
children_values
```

```python
smoker_values = insurance_data['smoker'].value_counts()
smoker_values.plot(kind='bar')
plt.xlabel('smoker')
plt.ylabel('count')
plt.title('Distribution of smokers')
plt.show()
```

```python
smoker_values
```

```python
region_values = insurance_data['region'].value_counts()
region_values.plot(kind='bar')
plt.xlabel('region')
plt.ylabel('count')
plt.title('Distribution of region')
plt.show()
```

```python
region_values
```

```python
insurance_data['charges'].hist(bins = 100)
plt.xlabel('charges')
plt.ylabel('Frequency')
plt.title('Histogram of charges')
plt.show()
```

```python
insurance_data.replace({'sex':{'male':0, 'female':1}}, inplace = True)
insurance_data.replace({'smoker': {'yes': 0, 'no':1}}, inplace = True)
insurance_data.replace({'region': {'southeast':0, 'southwest':1, 'northeast':2, 'northwest':3}}, inplace = True)
```
